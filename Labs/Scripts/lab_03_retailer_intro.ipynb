## Prepare Data

➡️ Assignment Tasks
- Create a "revenue" variable and make sure it is a numeric datatype
- Remove rows with empty (NULL) values
- Locate outliers in at least two of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
df['revenue'] = df['quantity'] * df['product_price']
df.head()
df.info()
```

```python
df
```

```python
# here I will drop any null values for quantity
df = df.dropna(subset=['quantity'])
# next I will drop any rows where the order is before 1990 and after 2025
def filter_year(date_str):
    try:
        year = int(date_str[:4])
        return 1990 <= year <= 2025
    except (ValueError, TypeError):
        return False

df = df[df['order_date'].apply(filter_year)]
```

```python
# dropping the null values
df = df.dropna()
df[df.isnull().any(axis=1)]
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per product line
- Display the total revenue for all months

```python
#average price
average_price_per_product_line = df.groupby('product_line')['product_price'].mean()
average_price_per_product_line
```

```python
#total revenue
total_revenue_by_month = df.groupby('order_year_month')['revenue'].sum()
total_revenue_by_month
```
